```python
import requests

def get_wikipedia_page_text(page_title):
    url = f"https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "prop": "extracts",
        "format": "json",
        "explaintext": True,
        "titles": page_title
    }
    response = requests.get(url, params=params)
    data = response.json()
    page = next(iter(data['query']['pages'].values()))
    return page['extract']

```

```python
import requests

def get_wikipedia_page_images(page_title):
    url = "https://en.wikipedia.org/w/api.php"

    # Step 1: Get image file names
    images_params = {
        "action": "query",
        "prop": "images",
        "format": "json",
        "titles": page_title
    }
    response = requests.get(url, params=images_params)
    data = response.json()
    page = next(iter(data['query']['pages'].values()))

    if 'images' not in page:
        return []

    image_filenames = [image['title'] for image in page['images']]

    # Step 2: Get image URLs and captions
    images_info = []
    for image_filename in image_filenames:
        imageinfo_params = {
            "action": "query",
            "prop": "imageinfo",
            "format": "json",
            "titles": image_filename,
            "iiprop": "url|extmetadata"
        }
        response = requests.get(url, params=imageinfo_params)
        data = response.json()
        page = next(iter(data['query']['pages'].values()))
        if 'imageinfo' in page:
            image_url = page['imageinfo'][0]['url']
            caption = page['imageinfo'][0].get('extmetadata', {}).get('ImageDescription', {}).get('value', '')

            images_info.append({
                'url': image_url,
                'caption': caption
            })

    return images_info
```

```python
# Example usage:
page_title = "United States"
text = get_wikipedia_page_text(page_title)
print(text)
print("\n\n****************\n****************\n****************\n       ****************  TEXT Completed *****************       \n\n****************\n****************\n****************\n\n")
images_info = get_wikipedia_page_images(page_title)
for info in images_info:
    print(f"URL: {info['url']}")
    print(f"Caption: {info['caption']}\n")
```
